```python
def countPoints(f):
    if f["name"] == "Tulsa" or f["geojson"]["type"] != "Polygon": # Tulsa can be a MultiPolygon
        return -10
    elif len(f["geojson"]["coordinates"][0]) < 10:
        return -10
    elif f["name"] == "Fresno":
        return len(f["geojson"]["coordinates"][0])

    elif f["addresstype"] != "city":
            return -1
    else:
        return len(f["geojson"]["coordinates"][0])


def selectIdx(res):
    """Select the first county level if found multiple entries"""
    rs = res.json()
    ps = [countPoints(x) for x in rs]
    return ps.index(max(ps))


def queryFeature(query):
    """Hit the GET endpoint and parse feature"""
    url = "https://nominatim.openstreetmap.org/search.php?q=" + query + "&polygon_geojson=1&format=jsonv2"
    res = requests.get(url)
    if len(res.json()) > 0:
        idx = selectIdx(res)
        j = res.json()[idx]
    else:
        j = res.json()[0]
    features = []
    f = { "type": "Feature"}
    p = {}
    p["place_id"] = j["place_id"]
    p["licence"] = j["place_id"]
    p["osm_type"] = j["osm_type"]
    p["osm_id"] = j["osm_id"]
    p["lat"] = j["lat"]
    p["lon"] = j["lon"]
    p["category"] = j["category"]
    p["type"] = j["type"]
    p["place_rank"] = j["place_rank"]
    p["importance"] = j["importance"]
    p["addresstype"] = j["addresstype"]
    p["name"] = j["name"]
    p["display_name"] = j["display_name"]
    f["geometry"] = j["geojson"]
    f["properties"] = p
    return f
```

```python
f = queryFeature("tulsa")
```

```python
fc = {"type": "FeatureCollection"}
fs = []
for c in tqdm(cities):
    print(f'processing {c}')
    f = queryFeature(c)
    fs.append(f)
fc["features"] = fs
```

```python
with open("osm-boundaries.geojson", "w") as file:
    json.dump(fc, file)
```
