```python
# init
qc = QuantumCircuit(4,2)
# qc.x(0)
qc.x(1)
# qc.x(2)

qc.h(0)
# qc.h(1)
# qc.h(2)
```

```python

qc.toffoli(0,1,3)
qc.cnot(0,1)
qc.toffoli(1,2,3)
qc.cnot(1,2)
qc.cnot(0,1)
qc.measure([2,3],[0,1])

# q0 and q1 are input variables, q2 is output, q3 is carryout
qc.draw(output="mpl")
```

```python
qsim = Aer.get_backend("aer_simulator")
job = execute(qc, backend=qsim, shots=1024)
result = job.result()

counts = result.get_counts(qc)

# right indexing - q0 is right most
display(plot_histogram(counts))
```
